## We can calculate the average salary of each job designation


```python
#We are creating a copy of the original dataframe so that we can make changes in the copy
# and at last we can merge the changes by merging both the data frames together
df_copy=df.copy()
```

```python
df_copy.drop('Job Description',axis=1,inplace=True)
```

```python
df_copy.head()
```

```python
df_copy['Salary Estimate']=df_copy["Salary Estimate"].str.split("(").str[0]
```

```python
df_copy["Salary Estimate"]
```

Created two new columns for range estimation of Salary "Starting Salary" and "End Salary". These two columns would be helpful later on in the calculation of the average salary

```python
df_copy["Starting Salary"]=df_copy["Salary Estimate"].str.split("-").str[0]
df_copy["End Salary"]=df_copy["Salary Estimate"].str.split("-").str[1]
print(df_copy[["Starting Salary","End Salary"]])
```

```python
df_copy["Starting Salary"]=df_copy["Starting Salary"].str.replace("$","")
df_copy["End Salary"]=df_copy["End Salary"].str.replace("$","")
df_copy["Starting Salary"]=df_copy["Starting Salary"].str.replace("K","000")
df_copy["End Salary"]=df_copy["End Salary"].str.replace("K","000")
```

```python
df_copy[["Starting Salary","End Salary"]]
```

```python
# Converting both of the columns into integer type so that at last we can compute the average salary 
df_copy["Starting Salary"]=df_copy["Starting Salary"].astype(int)
df_copy["End Salary"]=df_copy["End Salary"].astype(int)
```

```python
#Dropping the salary column
df_copy.drop("Salary Estimate", axis=1, inplace=True)
```

```python
df_copy.head()
```

```python
df_copy.info()
```

# we see that there is a negative value of the rating and it is not possible that the rating can be a negative number 

```python
#Checking how many values are negative and how many are the unique value count
df_copy["Rating"].value_counts()
```

```python
#Making the ratings back to positive
df_copy["Rating"]=df_copy["Rating"].abs()
```

```python
#We see anomaly in the name of the companies
df_copy["Company Name"].value_counts()
```

```python
#correcting the companies name
# we encountered one problem with the backslash as it is an escape character so we first
# replaced it forward slash and then removed it
df_copy["Company Name"]=df_copy["Company Name"].str.replace("\n","/",regex=True)
df_copy["Company Name"]=df_copy["Company Name"].str.split("/").str[0]
```
